## **Dataset Columns**
**The dataset consists of the following columns:**

1. **Person ID**: An identifier for each individual in the dataset.
2. **Gender**: The gender of the person (Male/Female).
3. **Age**: The age of the person in years.
4. **Occupation**: The occupation or profession of the person.
5. **Sleep Duration (hours)**: The number of hours the person sleeps per day.
6. **Quality of Sleep (scale: 1-10)**: A subjective rating of the quality of sleep, ranging from 1 to 10
7. **Physical Activity Level (minutes/day)**: The number of minutes the person engages in physical activity daily.
8. **Stress Level (scale: 1-10)**: A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
9. **BMI Category**: The BMI category of the person (e.g., Underweight, Normal, Overweight).
10. **Blood Pressure (systolic/diastolic)**: The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
11. **Heart Rate (bpm)**: The resting heart rate of the person in beats per minute.
12. **Daily Steps**: The number of steps the person takes per day.
13. **Sleep Disorder**: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

# **1.What is the distribution of sleep durations among different age groups?**


```python
df.head()
```

```python
print(df["Age"].min())
print(df["Age"].max())
```

```python
# Define the age ranges and corresponding class labels
age_ranges = [(0, 18), (19, 29), (30, 39), (40, 49), (50, 59), (60, 100)]
class_labels = ['0-18', '19-29', '30-39', '40-49', '50-59', '60+']


df['Age Class'] = pd.cut(df['Age'], bins=[r[0] for r in age_ranges] + [age_ranges[-1][1]], 
                         labels=class_labels, include_lowest=True)
```

```python
# Filter the DataFrame for each age class and create separate distribution plots
age_classes = ['19-29', '30-39', '40-49', '50-59']

plt.figure(figsize=(12, 8))
for i, age_class in enumerate(age_classes):
    plt.subplot(2, 3, i+1)
    sns.kdeplot(data=df[df['Age Class'] == age_class], x='Sleep Duration',hue=df["Sleep Disorder"])
    plt.xlabel('Sleep Duration (hours)')
    plt.ylabel('Density')
    plt.title(f'Distribution of Sleep Durations\n(Age Class: {age_class})')

plt.tight_layout()
plt.show()
```

# **3. Are there any differences in sleep disorders based on occupation?**

# **4. What is the average heart rate for individuals with different sleep disorders?**


```python

average_heart_rate = df.groupby('Sleep Disorder')['Heart Rate'].mean().reset_index()


average_heart_rate_table = pd.DataFrame(average_heart_rate)
average_heart_rate_table.columns = ['Sleep Disorder', 'Average Heart Rate']
average_heart_rate_table = average_heart_rate_table.round({'Average Heart Rate': 2})
average_heart_rate_table = average_heart_rate_table.sort_values(by='Average Heart Rate', ascending=False)


table = tabulate(average_heart_rate_table, headers='keys', tablefmt='fancy_grid')

# Display the formatted table
print(table)
```

# **9. How does blood pressure vary based on sleep duration?**

```python
print(df["Blood Pressure"].min())
print(df["Blood Pressure"].max())
```

```python
df["Blood Pressure"].unique()
```

```python

low_threshold = 90
high_threshold = 120


def map_blood_pressure(bp):
    systolic, diastolic = map(int, bp.split('/'))
    if systolic < low_threshold and diastolic < low_threshold:
        return 'Low'
    elif systolic >= high_threshold or diastolic >= high_threshold:
        return 'High'
    else:
        return 'Normal'

# Apply the blood pressure mapping function to the 'Blood Pressure' column
df['Blood Pressure Class'] = df['Blood Pressure'].apply(map_blood_pressure)
```

```python
# Create a box plot to visualize the variation of blood pressure based on sleep duration
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sleep Duration', y='Blood Pressure Class')
plt.title('Blood Pressure Variation based on Sleep Duration')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Blood Pressure Class')
plt.show()
```

# **10. What is the most common sleep disorder among different age groups?**

```python
df.head()
```

```python

most_common_sleep_disorder = df.groupby('Age Class')['Sleep Disorder'].apply(lambda x: x.value_counts().idxmax() if len(x) > 0 else 'No Data').reset_index()
most_common_sleep_disorder.columns = ['Age Class', 'Most Common Sleep Disorder']


table = tabulate(most_common_sleep_disorder, headers='keys', tablefmt='fancy_grid')


print(table)
```

#### People between age 40-49 suffer more from sleep disorder with isomnia being more prevalent
